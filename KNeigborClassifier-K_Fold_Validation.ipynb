```python
# Import numpy and pandas
import numpy as np
import pandas as pd

# Import operating specific functionality
import os

# Import Seaborn for visualisation
import seaborn as sn
# Import matplotlib for visualisation
import matplotlib.pyplot as plt

# Import timer
from timeit import default_timer as timer
```

```python
# Create variable "data" containing the data from the voice.csv file
data = pd.read_csv("voice.csv")
```

```python
# Show the head of the data, confirming that the data is read and that we have all columns
data.head()
```

```python
# Convert the label from text male/female to binary value of 1 or 0
data.label = [1 if each == "male" else 0 for each in  data.label]
# Show head of data to confirm that label value is changed
data.head()
```

```python
# Set variable "gender" to be the label column
gender = data.label
# Drop the label column from the data so that we now have features and labels separated
data.drop(["label"], axis = 1, inplace = True)
```

```python
#Normalize the data to ensure that differing magnitude within features does not create a problem when running the model
data = (data-np.min(data))/(np.max(data)-np.min(data))

# Set variable "features" to be the normalized data
features = data

features.head()
```

```python
# Import train_test_split function from scikit-learn
from sklearn.model_selection import train_test_split

# Split data into training set and testing set with a test size of 20% of the samples.
# The integer 42 is used for shuffling the data
x_train, x_test, y_train, y_test = train_test_split(features, gender, test_size = 0.2, random_state=42, stratify=gender)
```

```python
# Import cross validation
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
# Import KNN
from sklearn.neighbors import KNeighborsClassifier
# Import confusion matrix
from sklearn.metrics import confusion_matrix

K_Optimal = 0
best_youden = 0
youden_scores = []
cv_scores = []

for K in range(1, 16):
    
    print("Cross validation scores for K of {}: {}".format(K, scores))

    # Create a KNN model
    knn = KNeighborsClassifier(n_neighbors = K)
    
    # Train the  model with a cross validation value of 5
    y_train_pred = cross_val_predict(knn, x_train, y_train, cv = 5)
    
    # Get the score for the cross validation value of 5
    kfold_score_start = timer()
    scores = cross_val_score(knn, x_train, y_train, cv = 5)
    print("K-fold score mean: {}".format(scores.mean()))
    kfold_score_end = timer()
    # Get the time for cross validation
    kfold_score_time = (kfold_score_end - kfold_score_start)
    
    # Print a simple confusion matrix for this cross validation
    print("Confusion Matrix for K of {}".format(K))
    print(confusion_matrix(y_train, y_train_pred))
    
    # Get the values for True Negatives,False Positives, Falses Negatives and True Positives
    tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()
    # Print the 4 values from above
    print("tn: {}, fp: {}, fn: {}, tp: {}".format(tn, fp, fn, tp))
    
    # Calculate sensitivity and Specificity
    sens = tp / (tp + fn)
    spec = tn / (fp + tn)
    
    # Calculate the Youdens Index and add it to the array of all youden indeces
    youden = sens - (1 - spec)
    youden_scores.append(youden)
    
    # Add the score for this cross validation to the array of scores
    cv_scores.append(scores)
    
    print("Accuracy of score of K of {}: {:.4f}(+/- {:.4f})".format(K, scores.mean(), scores.std() * 2))
    print("Youdens Index for K of {}: {:.4f}\n ".format(K, youden))
    
    # Check if the Youdens Index is the best and if so set this value and the optimal K value
    if best_youden < youden:
        best_youden = youden
        K_Optimal = K
        

# Print the optimal K value and it's accuracy
print("The optimal K value is {} with a Youden Index of {:.4f}".format(K_Optimal, best_youden))
```

```python
# Now set K to 4 neighbours
K = 4
# Set variable "knn" to be the KNeighborsClassifier using K of 3 from above
knn_fin = KNeighborsClassifier(n_neighbors = K)

# Train the model using the training sets
knn_fin.fit(x_train, y_train)

# Predict the response for the test dataset
y_pred = knn_fin.predict(x_test)

# Import metrics module from scikit-learn
from sklearn import metrics
# Check how often the knn model is correct
accuracy = metrics.accuracy_score(y_test, y_pred)
# Print the result
print("Accuracy of model was: ", accuracy)

# Get values for true positive (true male), true negative (true female),
# false positive (false male) and false negative (false female)
tp = sum((y_test == y_pred) * (y_test == 1) * 1)
print("TP: ", tp)
tn = sum((y_test == y_pred) * (y_test == 0) * 1)
print("TN: ", tn)
fp = sum((y_test != y_pred) * (y_test == 0) * 1)
print("FP: ", fp)
fn = sum((y_test != y_pred) * (y_test == 1) * 1)
print("FN: ", fn)

# Define the sensitivity and specificity
sens = tp / (tp + fn)
spec = tn / (fp + tn)

print("Sensitivity is {}".format(sens))
print("Specificity is {}".format(spec))

# Calculate the Youden Index
youden = sens - (1 - spec)

print("Youden Index is {}".format(youden))
```

```python
# Set dataset for confusion matrix using the test and prediction data
confusion_data = [[tn, fp, tn+fp], [fn, tp, fn+tp], [tn+fn, fp+tp, tn+fp+fn+tp]]

pd.options.display.float_format = '{:.4f}'.format

# Define the Pandas dataframe
df = pd.DataFrame(confusion_data, ['Actual Female', 'Actual Male', 'Total'], ['Predicted Female', 'Predicted Male', 'Total'])
sn.set(font_scale=1.4)
sn.heatmap(df, annot=True, fmt='g')

plt.title('KNeighborsClassifier Confusion Matrix', loc='center')
plt.show()
```
