```python
# Import division to ensure that division returns a float
from __future__ import division
# Import numpy and pandas
import numpy as np
import pandas as pd
# Import operating specific functionality
import os
```

```python
# Create variable "data" containing the data from the voice.csv file
data = pd.read_csv("voice.csv")
```

```python
# Show the head of the data, confirming that the data is read and that we have all columns
data.head()
```

```python
# Convert the label from text male/female to binary value of 1 or 0
data.label = [1 if each == "male" else 0 for each in  data.label]
# Show head of data to confirm that label value is changed
data.head()
```

```python
# Set variable "gender" to be the label column
gender = data.label
# Drop the label column from the data so that we now have features and labels separated
data.drop(["label"], axis = 1, inplace = True)
```

```python
#Normalize the data to ensure that differing magnitude within features does not create a problem when running the model
data = (data-np.min(data))/(np.max(data)-np.min(data))

# Set variable "features" to be the normalized data
features = data

features.head()
```

```python
# Import train_test_split function from scikit-learn
from sklearn.model_selection import train_test_split

# Split data into training set and testing set with a test size of 20% of the samples.
# The integer 42 is used for shuffling the data
x_train, x_test, y_train, y_test = train_test_split(features, gender, test_size = 0.2, random_state = 42, stratify=gender)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
import time
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix



F_Optimal = 0
best_youden = 0
youden_scores = []
cv_scores = []


for F in range(1,20):
    
    RFC = RandomForestClassifier(n_estimators=100, max_features=F)
    
    y_train_pred = cross_val_predict(RFC, x_train, y_train, cv = 5)
    
    scores= cross_val_score(RFC,x_train,y_train, cv= 5)
    
    print("Scores")
    print(scores)

    print("Confusion Matrix for F (max_features) of {}".format(F))
    print(confusion_matrix(y_train, y_train_pred))
    
    # Get the values for True Negatives,False Positives, Falses Negatives and True Positives
    tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()
    # Print the 4 values from above
    print("tn: {}, fp: {}, fn: {}, tp: {}".format(tn, fp, fn, tp))
    
    # Calculate sensitivity and Specificity
    sens = tp / (tp + fn)
    spec = tn / (fp + tn)
    
    # Calculate the Youdens Index and add it to the array of all youden indeces
    youden = sens - (1 - spec)
    youden_scores.append(youden)
    
    # Add the score for this cross validation to the array of scores
    cv_scores.append(scores)
                     
    print("Accuracy of score of F of {}: {:.4f}(+/- {:.4f})".format(F, scores.mean(), scores.std() * 2))
    print("Youdens Index for F of {}: {:.4f}\n ".format(F, youden))
            
    if best_youden < youden:
        best_youden = youden
        F_Optimal = F

# Print the optimal F value and it's accuracy
print("The optimal F value is {} with a Youden Index of {:.4f}".format(F_Optimal, best_youden))


```
